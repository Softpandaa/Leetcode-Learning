```python
    # 70 Climbing Stairs
    # Idea: #ways to climb n = #ways to climb n-1 + #ways to climb n-2
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n<=2: return n
        arr = [0]*(n+1)
        arr[1]=1
        arr[2]=2
        for i in range(3,n+1):
            arr[i] = arr[i-1]+arr[i-2]
        return arr[n]
```

```python
    # 746 Min Cost Climbing Stairs
    # Idea: Cost to climb n= min(Cost to climb from n-1, Cost to climb from n-2)
    def minCostClimbingStairs(self, cost):
        """
        :type cost: List[int]
        :rtype: int
        """
        n = len(cost)
        dp = [0]*n
        dp[0] = cost[0]
        if n>=2:
            dp[1] = cost[1]
            for i in range(2, n):
                dp[i] = cost[i] + min(dp[i-1], dp[i-2])
        return min(dp[-1], dp[-2])
        
```

```python
    # 198 House Robber
    """Idea:
    Rob n = NoRob n-1 and Rob n
    No rob n = NoRob n-1 or NoRob n-2 = max(noRob, rob)
    Interchange Rob and noRob as no adj rob is allowed
    """
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        rob, norob = 0, 0
        for num in nums:
            newRob = norob + num
            newNoRob = max(norob, rob)
            rob, norob = newRob, newNoRob
        return max(rob, norob)
```

```python
    # 213 House Robber II
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # Result = max(subarray of [0:n-1], [1:n])
        if len(nums)==1: return nums[0]
        def helper(nums):
            rob1, rob2 = 0,0
            for num in nums:
                newRob = max(rob1+num, rob2)
                rob1 = rob2
                rob2 = newRob
            return rob2
        return max(helper(nums[1:]), helper(nums[:-1]))
```
