```python
    # 704 Binary Search
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        l, r = 0, len(nums)

        while l < r:
            m = l + ((r - l) // 2)  
            if nums[m] >= target:
                r = m
            elif nums[m] < target:
                l = m + 1
        return l if (l < len(nums) and nums[l] == target) else -1
```

```python
    # 74 Search a 2D Matrix
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        rows, cols = len(matrix), len(matrix[0])
        l,r = 0, rows*cols-1
        while l<=r:
            m = (l+r) // 2
            row, col = m // cols, m % cols
            current = matrix[row][col]
            if current>target: r=m-1
            elif current<target: l=m+1
            else: return True
        return False
```

```python
    # 875 Koko Eating Bananas
    """Idea: Count the hours by ceil(1.0*_)"""
    def minEatingSpeed(self, piles, h):
        """
        :type piles: List[int]
        :type h: int
        :rtype: int
        """
        l, r = 1, max(piles)

        while l <= r:
            k = (l + r) // 2
            totalTime = sum(math.ceil(1.0 * pile / k) for pile in piles) # Faster "for loop"
            if totalTime <= h: r = k - 1
            else: l = k + 1
        return l
```

```python
    # 4 Median of two sorted arrays
    """Idea: 
    1) Always keep B a lower array than A
    2) i = (l+r)//2, and A[:i]+B[:j]=half so j=haf-i-2 (Python starts with index 0)
    3) max from the left partitions of both arrays <= min from the right partitions of both the arrays
    4) Aleft>Bright means that i should decrease -> r = i-1
    5) total = odd -> min(RHS)
    6) total = even -> avg( max(LHR), min(RHS) )
    """
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """
        A, B = nums1, nums2
        total = len(nums1) + len(nums2)
        half = total // 2
        if len(A)>len(B): 
            A,B=B,A

        l,r=0,len(A)-1
        while True:
            i = (l+r)//2
            j = half - i - 2

            Aleft = A[i] if i >= 0 else float("-infinity")
            Aright = A[i + 1] if (i + 1) < len(A) else float("infinity")
            Bleft = B[j] if j >= 0 else float("-infinity")
            Bright = B[j + 1] if (j + 1) < len(B) else float("infinity")
            
            if Aleft<=Bright and Bleft<=Aright:
                if total % 2: return min(Aright,Bright)
                return float(max(Aleft,Bleft)+min(Aright,Bright))/2
            elif Aleft>Bright: r=i-1
            else: l=i+1
```

```python
# 2034
"""Idea:
1) Store price[timestamp] and last_t
2) maxHeap: -p; minHeap: p => Get the (-max, min)
3) heappush: Push value item on the heap
4) heappop: Pop smallest item from the heap
"""
class StockPrice(object):

    def __init__(self):
        self.price_at = {}
        self.last_t = 0
        self.minHeap = []
        self.maxHeap = []

    def update(self, timestamp, price):
        """
        :type timestamp: int
        :type price: int
        :rtype: None
        """
        self.price_at[timestamp] = price
        self.last_t = max(self.last_t, timestamp)
        heappush(self.minHeap, (price, timestamp))
        heappush(self.maxHeap, (-price, timestamp))

    def current(self):
        """
        :rtype: int
        """
        return self.price_at[self.last_t]

    def maximum(self):
        """
        :rtype: int
        """
        while True:
            p, t = self.maxHeap[0]
            if self.price_at[t] == -p: return -p
            heappop(self.maxHeap)
        

    def minimum(self):
        """
        :rtype: int
        """
        while True:
            p, t = self.minHeap[0]
            if self.price_at[t] == p: return p
            heappop(self.minHeap)


# Your StockPrice object will be instantiated and called as such:
# obj = StockPrice()
# obj.update(timestamp,price)
# param_2 = obj.current()
# param_3 = obj.maximum()
# param_4 = obj.minimum()
```
