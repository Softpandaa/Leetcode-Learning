```python
    # 121 Best Time to Buy and Sell Stock
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        minn = prices[0]
        profit = 0
        for price in prices:
            if price < minn: minn = price
            profit = max(profit, price-minn)
        return profit
```

```python
    # 3 Longest substring without repeating characters
    """Idea:
    1) Create a set of seen charaters
    2) Move the window if see duplicate
    """
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        charS = set()
        l = 0
        res = 0
        for r in range(len(s)):
            while s[r] in charS:
                charS.remove(s[l])
                l += 1
            charS.add(s[r])
            res = max(res, r-l+1)
        return res
```

```python
    # 424 Longest Repeating Character replacement
    """Idea: Create a moving window that check the max char is valid or not
    Valid condition: Over total length of n, with the max frequent char m, we need n-m>=k so as we can only replace k char
    If current window is valid, then move the RHS +1
    Else, move the window to right by 1
    """
    def characterReplacement(self, s, k):
        """
        :type s: str
        :type k: int
        :rtype: int
        """
        count = {}
        res = 0
        l = 0
        maxf = 0
        for r in range(len(s)):
            count[s[r]] = 1 + count.get(s[r],0)
            maxf = max(maxf, count[s[r]])
            while (r-l+1) - maxf > k:
                count[s[l]] -= 1
                l += 1
            res = max(res, r-l+1)
        return res
```

```python
    # 567 Permutation in String
    """Idea:
    Create a matches table that get frequency of "a-z", if matches=26 then current is a permutation of s1
    Calculate frequency of current for each moving windoe is inefficient
    Initiate the matches to be that of the frequency of s1 and s2
    Move to the next charater, adjust +1 or -1; Then check left character, adjust +1 or -1
    Move the window
    """
    def checkInclusion(self, s1, s2):
        """
        :type s1: str
        :type s2: str
        :rtype: bool
        """
        s1_len = len(s1); s2_len = len(s2)
        if s1_len>s2_len: return False


        s1_count, s2_count = [0]*26, [0]*26
        for i in range(s1_len):
            s1_count[ord(s1[i])-ord('a')] += 1
            s2_count[ord(s2[i])-ord('a')] += 1

        matches = sum([1 if s1_count[i]==s2_count[i] else 0 for i in range(26)])

        l = 0
        for r in range(s1_len, s2_len):
            if matches == 26: return True

            next_idx = ord(s2[r])-ord('a')
            s2_count[next_idx] += 1
            if s1_count[next_idx] == s2_count[next_idx]: matches += 1
            elif s1_count[next_idx]+1 == s2_count[next_idx]: matches -= 1

            current_idx = ord(s2[l])-ord('a')
            s2_count[current_idx] -= 1
            if s1_count[current_idx] == s2_count[current_idx]: matches += 1
            elif s1_count[current_idx]-1 == s2_count[current_idx]: matches -= 1

            l += 1
        
        return matches == 26
```

```python
    # 76 Minimum window string
    """Idea: 
    Create a count for t, 
    Find the minimum substring that fulfills the count of t
    Move the left pointer to find the minimum substring
    """
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        if len(t)>len(s): return ""

        t_count = {}
        for char in t: t_count[char] = 1+t_count.get(char,0)
        matches_goal = len(t_count)

        l = 0; s_count = {}
        matches = 0
        res_l, res_r = -1e5, float("infinity")
        for r in range(len(s)):
            current_char = s[r]
            s_count[current_char] = 1+s_count.get(current_char,0)
            if current_char in t_count and t_count[current_char]==s_count[current_char]: matches += 1
            while matches == matches_goal:
                if (r-l+1) < (res_r-res_l+1): res_l, res_r = l, r
                current_lchar = s[l]
                s_count[current_lchar] -= 1
                if current_lchar in t_count and s_count[current_lchar] < t_count[current_lchar]: matches -= 1
                l += 1
        if (res_r-res_l+1) != float("infinity"): return s[res_l: res_r+1]
        else: return ""
```

```python
    # 239 Sliding window maximum
    """Idea:
    Use q=deque(): allows insertion and deletion at both ends
    q contains the index of current window, but maximum's index at the beginning
    Method: 
    1) If add > current, remove index of current
    2) Ensure the window's index is valid
    """
    def maxSlidingWindow(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: List[int]
        """
        output = []
        q = deque()
        l = r = 0
        while r < len(nums):
            while q and nums[q[-1]] < nums[r]: q.pop()
            q.append(r)
            if l > q[0]: q.popleft()
            if (r+1)>=k:
                output.append(nums[q[0]])
                l += 1
            r += 1
        return output
        
```
